**Q1: Simulate aliasing of sinusoidal signals**
For this problem you will simulate the phenomena of aliasing. This will give you pracitice in making synthetic time series, subsetting data, and plotting your results.
1 To start, make a time vector that goes from 0 to 120 s, with 10000 entries in it. Then make a synthetic time series that has two close in frequency sine waves:
$$x=A1cos(ωt)+A2cos((ω+δω)t)$$
where $ω$ is the "carrier" frequency, and $δω<<ω$ is the difference in frequencies between the two waves. For simplicty, you could make $A1=A2=1.[V]$. Feel free to choose an interesting $ω$.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
%matplotlib inline
```

```python
#set initial values
T = np.linspace(0, 120, 10000)
a1 = 1
a2 = 1
w = .5
dw = w/10

# set values for components of wave
def setvalues(dw):
    x1, x2, X = [], [], []
    for t in T:
        x1.append(a1*np.cos(w*t))
        x2.append(a2*np.cos((w + dw)*t))
        X.append(x1[len(x1)-1] + x2[len(x2) - 1])
    return x1, x2, X

x1, x2, X = setvalues(dw)
```

2 Now plot the two component waves and the full value of x as a time series. Preferably on the same axis. Make sure to label your axes and include a legend

```python
#makes plot of waves
def makeplot(T,X,x1,x2):
    
    #plot each wave
    plt.plot(T, X, 'r', label = "$A_1cos(\omega t) + A_2cos(\omega + \delta \omega)t)$")
    plt.plot(T, x1,'b', label = "$A_1cos(\omega t)$")
    plt.plot(T, x2,'g', label = "$A_2cos(\omega + \delta \omega)t)$")
    
    #label axis
    plt.ylabel('Position')
    plt.xlabel('Time (seconds)')
    plt.title("$\omega$ = {}, $\delta \omega$ = {}".format(w, dw))
    
    plt.legend(bbox_to_anchor=(1.6,1))
    plt.show()
    
makeplot(T,X,x1,x2)
```

3 Try a few different values of δω, and compare.

```python
#makes multiple plots with different delta omegas
for n in [20, 50, 100]:
    dw = w/n
    x1, x2, X = setvalues(dw)
    makeplot(T,X,x1,x2)
```

**Q2: Demonstrate the central limit theorem for a very unfair coin flip.**
Consider a coin that is twice as likely to come up heads as tails. To put a number on it, call tails $x=−1/2$ and heads $x=+1/2$.

**1** What are the mean, $μ_x$, and variance $σ^2_x$ of this process?

```python
# finds average value based on probability of each result
def getmean(prob, result):
    return np.sum(prob*result)
 
# finds variance 
def getvar(prob, result, mean):
    x = (result - mean)**2
    return np.sum(x*prob)

coinprob = np.array([1./3., 2./3.])
coinresult = np.array([-0.5, 0.5])

mean = getmean(coinprob, coinresult)
print "The mean value of the coin flip is",mean

var = getvar(coinprob, coinresult, mean)
print "The variance of the coin flip is", var
```

**2** Plot or represent the probability distribution (or cumulative distrubution) of this process.

```python
def makebar(res,prob):
    plt.bar(res-0.125, prob, 1/4.)
    plt.axis([-1, 1, 0, 0.7])
    plt.ylabel('Probability')
    plt.xlabel('Result')
    plt.title('Probability Distribution of an Unfair Coin Flip')
    plt.show()
    

    
makebar(coinresult, coinprob)
```

**3** Write a function using the np.random.rand function to simulate the coin flip. It shoudl return $+1/2$ twice as often as $−1/2$. Demonstrate that your function works.

```python
#coin flip simulation
def coinflip(flips):
    
    heads = 0
    tails = 0
    
    for x in range(0, flips):
        if np.random.rand() < 1./3.:
            tails = tails + 1
        else:
            heads = heads + 1
    
    return tails, heads

# testing that it works
flips = 100
tails, heads = coinflip(flips)
print "Out of {} flips there were {} heads and {} tails there {} as many heads as tails".format(flips, heads, tails, heads/float(tails))
```

**4** Perform Nflip and take the sample mean x¯¯¯ of the values. Do this 2000 times, and show that as $N_{flip}→∞$ the distribution of x¯¯¯ approaches a Normal distribution by plotting the histograms of the means for each. For $N_{flip}$ choose 2, 16, 64, 256.

Comment on how the convergence of x¯¯¯ for the unfair coin compares to the convergence of the sample mean of the dice roll from the notes.

```python
def multiflips(n):
    means = np.zeros(2000)
    count = 0
    while count < 2000:
        
        tails, heads = coinflip(n)
        
        tprob = tails/float(n)
        hprob = heads/float(n)
        
        probs = np.array([tprob, hprob])
        
        means[count] = getmean(probs, coinresult)
        count = count+1
    
    return means

for n in [2, 16, 64, 256]:
    
    means = multiflips(n)
        
    plt.hist(means, 40, histtype='step', label = '{} flips'.format(n))
    plt.title('DIstribution of means for multiple coin flips'.format(n))
    plt.ylabel('Occurences')
    plt.xlabel('Mean Value')
    
    mean = np.sum(means)/len(means)
    
plt.legend(bbox_to_anchor=(1.3,1))    
plt.show()
```

As the number of flips increases, The distribution of the means approaches a normal distribution.

When comparing the results with that of the dice, it can be seen the the coin flips produce a wider normal distribution with a lower peak and a higher standard deveation.
